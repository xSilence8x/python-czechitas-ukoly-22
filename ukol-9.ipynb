```python
import pandas
```

```python
teploty = pandas.read_csv("temperature.csv")
teploty
```

Dotaz na měření, která byla provedena v Praze.

```python
teploty[teploty["City"].isin(["Prague"])]
```

Dotaz na měření, ve kterých je teplota (sloupec AvgTemperature) vyšší než 80 stupňů.

```python
teploty[teploty["AvgTemperature"] > 80]
```

Dotaz na měření, ve kterých je teplota vyšší než 60 stupňů a současně bylo měření provedeno v regionu (sloupec Region) Evropa (Europe).

```python
teploty[(teploty["AvgTemperature"] > 60) & (teploty["Region"] == "Europe")]
```

Dotaz na extrémní hodnoty, tj. měření, ve kterých je teplota vyšší než 80 stupňů nebo menší než - 20 stupňů.

```python
teploty[(teploty["AvgTemperature"] > 80) | (teploty["AvgTemperature"] < -20)]

```

```python
import pytemperature
```

```python
nove_teploty = teploty
```

zkus si vytvořit nový sloupec, který bude obsahovat průměrnou templotu ve stupních Celsia

```python
nove_teploty["AvgTemperatureCelsia"] = pytemperature.f2c(teploty["AvgTemperature"])
nove_teploty
```

Dotaz na měření, ve kterých je teplota (sloupec AvgTemperatureCelsia) vyšší než 30 stupňů Celsia.

```python
nove_teploty[nove_teploty["AvgTemperatureCelsia"] > 30]
```

Dotaz na měření, ve kterých je teplota vyšší než 15 stupňů Celsia a současně bylo měření provedeno v regionu (sloupec Region) Evropa (Europe).

```python
nove_teploty[(nove_teploty["AvgTemperatureCelsia"] > 15) & (nove_teploty["Region"] == "Europe")]
```

Dotaz na extrémní hodnoty, tj. měření, ve kterých je teplota vyšší než 30 stupňů Celsia nebo menší než -10 stupňů.

```python
nove_teploty[(nove_teploty["AvgTemperatureCelsia"] > 30) | (nove_teploty["AvgTemperatureCelsia"] < -10)]
```
